2. Torre de Saruman

Saruman el Blanco es uno de los personajes de la novela El Señor de los Anillos, creada por J. R. R. Tolkien. Se trata del líder de los Istari, magos humanos, que en principio tiene la intención de vencer a Saurón, el Señor Oscuro, pero que termina deseando su poder para conquistar la Tierra Media.

Supongamos que existe un juego de mesa llamado "Torre de Saruman" que consiste en una serie de piezas en una pila, cada pieza tiene un valor entero positivo. Comenzando en la cima de la pila, si la suma del valor P de la última pieza apilada y el valor Q de la pieza justo debajo es un número par, ambas piezas de desapilan y en su lugar se apila una de tamaño (P + Q) / 2. De allí en adelante se sigue aplicando la misma regla hasta que posiblemente no quede sino una pieza.

Considerando lo anterior, y dada una secuencia de piezas en la pila, ¿harías un programa para determinar la altura final de la torre?

## Input
La entrada comienza con una línea que contiene la cantidad C de casos, no más de 50. Luego siguen C líneas, cada una con las piezas de la pila correspondiente. Cada pila tendrá por lo menos una y no más de 2000 piezas, las cuales tienen un valor entero positivo, no necesariamente diferente, inferior a 999999. La primera pieza es la que está en la base de la pila, la segunda la inmediatamente arriba y así sucesivamente. Dichos valores están separados entre sí por un espacio en blanco.

## Output
La salida debe tener C líneas, cada una con dos valores separados entre sí por un espacio en blanco: la altura final de la torre y el valor de la pieza en su cima para cada caso.

## Ejemplo
input
3 \n
5 4 3 2 1 31 \n
32 16 8 4 \n
3 2 1 \n

output
1 5\n
2 11\n
3 1\n




```python
import sys
from collections import deque
 
 
casos = int(sys.stdin.readline().strip()) #o(1)
 
for i in range(casos): #o(C)
    entrada = list(map(int, sys.stdin.readline().split()))#o(N)
    pila = deque(entrada)#o(N)
    #print(pila) 

    for j in range(len(pila)-1,0,-1): #o(N)
        p = pila.pop() #o(1)
        q = pila.pop() #o(1)
        if (p+q)%2 == 0: #o(1)
            pila.append((p+q)/2) #o(1)
        else:
            pila.append(q) #o(1)
            pila.append(p) #o(1)
            break
 
        #print(f"{p}-{q}") 
    print(f"{len(pila)} {int(pila[-1])}")#o(1)

#Al final la complejidad es O(C+N)
```

A partir del código anterior se tiene una complejidad de O(C+N) con C el número de casos y N el tamaño de la pila.
