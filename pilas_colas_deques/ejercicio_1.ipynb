1. Fila de revendedores
time limit per test1 second
memory limit per test256 megabytes

Donde se vendan boletas habrán revendedores, ¿no cierto? Pues bueno, nuestro problema consiste en que para un determinado evento hay una fila para comprar las boletas y solamente hay disponibles T. Los organizadores del evento creen que han encontrado una solución: podrán un cordón de seguridad, quien lo pase podrá hacer la fila y si sale de él no podrá ingresar de nuevo. El paso de dicho cordón estará restringido de 8: 00 a 10: 00am, tiempo después del cual se abrirá la taquilla. Además, a cada persona solo se le venderán hasta 6 boletas.

Sin embargo este plan tenía un fallo: alguien después de comprar podría no salir del cordón de seguridad y en cambio volver a hacer la fila. Considerando este potencial riesgo los organizadores planearon implementar una regla: después de comprar boletas (independiente de cuántas) no se podría volver a hacer la fila. El lio es que los revendedores se enteraron del plan e hicieron una protesta que casi termina haciendo cancelar el evento.

Allí, los organizadores llegaron con otra solución que apaciguó los ánimos: de cada cinco personas (no necesariamente diferentes) que compren, solo a la quinta se le vetará de hacer la fila de nuevo. Los cuatro restantes podrán volver a hacerla pero solo si compran en la siguiente ocasión la misma cantidad de boletas que cada una acabó de comprar. (o menos, si ya no hay disponibles suficientes)

## Input
La entrada comienza con una línea que contiene dos valores enteros positivos separados entre sí por un espacio en blanco: N (1 ≤ N ≤ 12000) que corresponde a la cantidad de compradores (todos revendedores) que lograron pasar el cordón de seguridad antes de las 10: 00am, y T (1 ≤ T ≤ 50000) que corresponde a la cantidad de boletas disponibles. Luego siguen N líneas, cada una con dos valores enteros positivos separados entre sí por un espacio en blanco: el número de identificación del revendedor R (1 ≤ C ≤ 9999999), y la cantidad K de boletas que va a comprar (1 ≤ K ≤ 6).

## Output
La salida debe tener una única línea con:

* Dos valores separados entre sí por un espacio en blanco: el número de identificación del vendedor que compra la última boleta y la cantidad de boletas que compró la última vez; o

* El mensaje (sin comillas) "quedaron boletas disponibles" en caso que la fila se agote antes de vender la totalidad de las boletas disponibles.

## ejemplo
Input
5 35
1001 3
1002 3
1003 3
1004 3
1005 3

Output
1003 2

```python
#solución
import sys
from collections import deque
 
ids = deque() #notese que los datos se comportan como 
#una cola
#nbs: numero de boletas
nbs = deque()

#N: num de compradores
#T: num de boletas
N, T= map(int, sys.stdin.readline().split()) #O(1)
 
#ciclo para leer todos los compradores con sus boletas
for i in range(N): #O(N)
    lineas = sys.stdin.readline().split() #lectura de datos
    id = lineas[0] #información 
    b = int(lineas[1])
    ids.append(id) #O(1)
    nbs.append(b) #O(1)
 
c = 1
while(len(ids)!=0): #O(N)
    id_persona = ids.popleft() #Primer persona en cola. #O(1)
    boleta_persona = nbs.popleft() #Numero de boletas. #O(1)
 
    #Si quedan menos de las que necesitan la última persona
    #las compra
    if T < boleta_persona: #O(1)
        break  #O(1)
    
    #se quita la cantidad de boletas compradas
    T -= boleta_persona #O(1)
 
    #print(f"{id_persona} {boleta_persona}")
    #print(T)
    #print(c)
    #if c%5 == 0:
    #    print(f"{ids} sin {id_persona}")
    #else:
    #    print(ids)
    #print()
    
    if c % 5 != 0: #si la posición no es múltiplo de 5. #O(1)
        ids.append(id_persona) #Entra el id. #O(1)
        nbs.append(boleta_persona) #Compra el mismo número de boletas. #O(1)
    c+=1
 
#print(f"T = {T}")
 
 
#si aún hay boletas pero no compradores
if T !=0 and len(ids)== 0:  #O(1) 
    print("quedaron boletas disponibles")
 
# Si quedan menos boletas de las que necesitan se tiene 
#a la última persona y las boletas restantes
else:  #O(1)
    print(f"{id_persona} {T}")

#por tanto el costo de este código es o(N) siendo N el número 
#de personas que harán la fila.
```
