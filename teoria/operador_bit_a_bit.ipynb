### El truco bitwise n & (n-1) == 0

Para entender por qué esto detecta una potencia exacta de 2 (2,4,8,16,32,...), se tiene que mirar cómo se ven los n´umero en el lenguaje nativo del procesador: **el sistema binario**.

Las potencias de 2 en binario tiene una característica única: TIENEN EXACTAMENTE UN SOLO BIT ENCENDIDO Y TODO EL RESTO SON CERO. 

¿Qué pasa si a una potencia de 2(n) le restamos 1?

Para n = 8:
- 8 en binario es 1000 (Un único uno al principio). Al restarle 1, que es 7 es 0111. 

Entonces, al aplicar la operación AND bit a bit (&)

8  ->  1 0 0 0
    7  ->  0 1 1 1
   ---------------
  8 & 7 -> 0 0 0 0  (Da exactamente 0)


  Para n = 12 que no es potencia de 2: 

  12  ->  1 1 0 0
   11  ->  1 0 1 1
   ---------------
 12 & 11 ->1 0 0 0  (Da 8, que NO es cero)

 Este truco solo sirve para ver que un número sea una potencia exacta de dos. Sin embargo, también el operador bit a bit es una forma bastante óptima de ver que un número es par. 

 Al hacer numero & 1 y el resultado es cero entocnes es un número par, de lo contrario es impar, por ejemplo: 

 12  ->  1 1 0 0
  1  ->  0 0 0 1
   ---------------
 12 & 1 ->0 0 0 0  (Da 0, es decir es par)

11  ->  1 0 1 1
1  ->  0 0 0 1
---------------
12 & 1->0 0 0 1  (Da 1, es decir impar.)


Nota: mirese que el operador bit a bit es como un tipo de conjunción: 
 ($A \land B$), pero con una diferencia clave: lo hace bit por bit sobre la representación binaria de los números.La regla es idéntica: Vale 1 únicamente si ambos bits son 1. Si hay algún 0 (o ambos son 0), el resultado es 0.
