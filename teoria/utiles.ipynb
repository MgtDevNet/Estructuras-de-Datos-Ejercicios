## ¿Por qué es una buena idea meter todo en una función solve() y usar if __name__ == 'main'?

En la mayoría de los lenguajes (como C++ o Java), usar una función main es obligatorio. En Python no lo es, pero en programación competitiva lo hacemos por dos razones de peso pluma (velocidad):

### **Razón A**: El gran secreto de Python (Las variables locales son más rápidas)

Cuando escribes código "suelto" en un script de Python (fuera de cualquier función), todas tus variables se crean en el ámbito global.

- Para leer o modificar una variable global, Python tiene que buscarla en un diccionario interno (globals()) usando su nombre en texto. Esto toma un pequeño tiempo extra en cada iteración.

- Cuando metes el código dentro de una función (def solve():), las variables se vuelven locales. El intérprete de Python (CPython) optimiza las variables locales guardándolas en un arreglo indexado de memoria de tamaño fijo en la RAM. Acceder a ellas es directo e increíblemente rápido.

El impacto: En bucles gigantescos ($10^5$ o $10^6$ iteraciones), correr el mismo código dentro de una función puede llegar a ser entre un 15% y un 30% más rápido que correrlo de forma global. Ese porcentaje es la diferencia entre un Accepted y un Time Limit Exceeded.

### Razón B: El estándar de ejecución (__main__)

La línea if __name__ == '__main__': le dice a Python: "Ejecuta este bloque de código solo si el archivo fue abierto directamente por el usuario (o por el juez en línea)".

Si otra función o script intentara importar tu archivo en un test automatizado, no se dispararía la solución por accidente. Es un estándar de ingeniería de software que mantiene tu plantilla limpia, organizada y lista para producción o competencia.
