## Búsqueda Binaria

La búsqueda binaria es uno de los algoritmos más importantes y eficientes en la informática. Su objetivo es encontrar la posición de un elemento dentro de un conjunto de datos en tiempo récord: $\mathcal{O}(\log n)$.

Para ponerlo en perspectiva: si buscas un número entre 1 millón de elementos, una búsqueda normal (recorrer uno por uno) tomaría hasta $1,000,000$ de pasos. Con búsqueda binaria, toma como máximo 20 pasos.

La Regla de Oro ⚠️
La búsqueda binaria SOLO funciona si la lista está previamente ORDENADA. Si la lista no está ordenada, el algoritmo fallará por completo.

¿Cómo funciona conceptualmente? (La estrategia de "Dividir y Vencerás")
Imagina que estás buscando una palabra en un diccionario impreso de 1000 páginas:

Abres el diccionario exactamente por la mitad (página 500).

Te fijas si la palabra que buscas está ahí, antes o después.

Si la palabra debería estar después, descartas de un solo golpe toda la primera mitad (páginas 1 a 500) y te quedas con la segunda mitad.

Repites el proceso abriendo a la mitad del nuevo tramo restante.

A cada paso descartas la mitad de los elementos que te quedan.

```python
# Idea de una implementación manual con python

def busqueda_binaria(lista, objetivo):
    izq = 0 #empieza en la posición 0
    der = len(lista) - 1 #empieza en la última posición

    while izq <= der:
        # Calculamos el punto medio exacto de la lista actual
        medio = (izq + der) // 2

        # Caso 1: ¡Lo encontramos!
        if lista[medio] == objetivo:
            return medio  # Retornamos el índice donde está

        # Caso 2: El objetivo es mayor, descartamos la mitad izquierda
        # es decir, ya se que mi objetivo está entre el punto
        #medio y todos los números a la derecha.
        elif lista[medio] < objetivo:
            izq = medio + 1

        # Caso 3: El objetivo es menor, descartamos la mitad derecha
        else:
            der = medio - 1

    # Esto va así sucesivamente hasta que suceden 2 cosas, o se encuentra
    #el número, o directamente no estaba en la lista. 
    return -1

    

# --- Ejemplo de uso ---
numeros = [10, 23, 35, 42, 50, 78, 90, 101] # Importante: Lista ordenada
objetivo = 78

resultado = busqueda_binaria(numeros, objetivo)

if resultado != -1:
    print(f"El número {objetivo} está en el índice: {resultado}")
else:
    print("El número no se encuentra en la lista.")
```

Ahora bien, para usarlo de una forma nativa desde el python se usa una librería llamada bisect. En Python no necesitas escribir el bucle while manualmente si no quieres. Python incluye una librería estándar ultra optimizada en C llamada bisect. Se tienen 2 funciones principales: 

**bisect_left y bisect_right**

**bisect_left(lista, x)**: Te dice en qué índice deberías insertar el número x para mantener la lista ordenada. Si x ya existe, te devuelve la primera posición (izquierda) donde aparece.

**bisect_right(lista, x)**: Similar, pero si x existe, te da la última posición (derecha).


```python
import bisect #librería

numeros = [10, 20, 30, 40, 50]

# Buscar dónde debería ir el número 35 para mantener el orden
indice = bisect.bisect_left(numeros, 35)
print(indice)  # Devuelve 3 (es decir, entre el 30 y el 40)

# Para verificar si un elemento existe usando bisect:
def existe(lista, objetivo):
    idx = bisect.bisect_left(lista, objetivo)
    # Comprobamos que el índice no sea mayor al tamaño y que el valor sea igual
    print(idx)
    return idx < len(lista) and lista[idx] == objetivo

print(existe(numeros, 30))  # el número está y es en la posición 2
print(existe(numeros, 25))  # el número no está PERO DEBERÍA
#ESTAR EN LA POSICIÓN 2

print(existe(numeros, 60)) #el número no está y dice que debería
#estar en la posición 5, y la posición máxima es 4. Por tanto, 
#el número no está. 

```

Hay que tener bastante cuidado con esta función, ya que si el valor no está en la lista aún así retornará una posición de donde debería ir el valor. Por tanto siempre que se use HAY QUE EVALUAR LA CONDICIÓN DE QUE EL ÍNDICE ESTE DENTRO DEL RANGO DE LA LSITA Y QUE EL VALOR DE LA POSICIÓN DEL ÍNDICE SI SEA EL OBJETIVO. 

La clave para entender la diferencia entre bisect_left y bisect_right en Python es pensar en qué lado de los elementos iguales se colocará el elemento que estás buscando si tuvieras que insertarlo en la lista para mantenerla ordenada.

bisect_left: Encuentra la posición de inserción A LA IZQUIERDA (antes) de cualquier elemento igual.

bisect_right: Encuentra la posición de inserción A LA DERECHA (después) de cualquier elemento igual.

1. Ejemplo 1: Caso Simple (Sin Repeticiones)Imagina que tenemos la lista:

x = [10, 20, 30, 40, 50]

Búsqueda de un valor existente ($30$):

bisect_left(x, 30) $\rightarrow$ Índice 2. Te dice: "A la izquierda del 30 está la posición 2".

bisect_right(x, 30) $\rightarrow$ Índice 3. Te dice: "A la derecha del 30 está la posición 3".

Búsqueda de un valor inexistente ($25$):

Cuando el número no existe en la lista, ambas funciones devuelven EXACTAMENTE LO MISMO:

bisect_left(x, 25) $\rightarrow$ Índice 2 (donde debería ir el 25 entre 20 y 30).

bisect_right(x, 25) $\rightarrow$ Índice 2 (exactamente la misma posición).

💡 Regla clave 1: Si el número no se repite o no existe, la única diferencia al buscar un número existente es que bisect_right te da un índice desplazado a +1.


```python
# Caso de las repeticiones
import bisect

x = [10, 20, 20, 20, 30]

idx_left = bisect.bisect_left(x, 20)
idx_right = bisect.bisect_right(x, 20)

print("bisect_left:", idx_left)   # Output: 1
print("bisect_right:", idx_right) # Output: 4
```

¿Qué nos dice cada resultado?
A) bisect_left nos dio el Índice 1
Apunta al PRIMER 20 de la lista.

Uso práctico: Te dice cuántos elementos son estrictamente MENORES que 20.
(En el índice 1, hay 1 elemento a la izquierda: el 10).

B) bisect_right nos dio el Índice 4
Apunta justo DESPUÉS DEL ÚLTIMO 20 (donde está el 30).

Uso práctico: Te dice cuántos elementos son MENORES O IGUALES a 20.
(En el índice 4, hay 4 elementos a la izquierda: 10, 20, 20, 20).
